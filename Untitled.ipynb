```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
nr = 100           # Number of spatial points
nt = 10000          # Number of time steps
r_max = 1.0        # Domain end
T = 10            # Final time
dr = r_max / (nr - 1)
dt = T / nt

# Spatial grid
r = np.linspace(0, r_max, nr)

# Initial condition
u = 1 - r**2
u_new = np.copy(u)

# Time stepping
for n in range(nt):
    # Compute diffusion coefficient
    D = np.exp(-u / 2.0)

    # Compute spatial derivatives
    dudr = np.zeros_like(u)
    dfluxdr = np.zeros_like(u)

    # Forward difference for dudr (central difference in the interior)
    dudr[1:-1] = (u[2:] - u[:-2]) / (2 * dr)
    dudr[0] = 0  # Neumann BC at r=0
    dudr[-1] = (u[-1] - u[-2]) / dr  # one-sided at r=1 for flux

    # Compute flux = r^2 * D * dudr
    flux = r**2 * D * dudr

    # Derivative of flux
    dfluxdr[1:-1] = (flux[2:] - flux[:-2]) / (2 * dr)
    dfluxdr[0] = (flux[1] - flux[0]) / dr  # one-sided
    dfluxdr[-1] = (flux[-1] - flux[-2]) / dr  # one-sided

    # Update u using the PDE
    u_new = u + dt * (1 / r**2) * dfluxdr

    # Enforce Dirichlet BC at r=1
    u_new[-1] = 0

    # Enforce Neumann BC at r=0 (∂u/∂r = 0 ⇒ symmetric about r=0)
    u_new[0] = u_new[1]

    # Prepare for next step
    u[:] = u_new

# Plot the final solution
plt.plot(r, u)
plt.xlabel("r")
plt.ylabel("u(r, T)")
plt.title("Non-linear diffusion at t = {:.2f}".format(T))
plt.grid(True)
plt.show()
```
